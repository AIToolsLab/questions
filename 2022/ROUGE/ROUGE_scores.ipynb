```python
# ROUGE for OpenAI [line by line scores]

files_rouge = FilesRouge()
scores_per_line_openai = files_rouge.get_scores("OpenAI_n-ROUGE.csv", "completion-reference-ROUGE.csv")
print(scores_per_line_openai)

# Extract only rouge-1 recall scores

with open("openai_rouge_1_recall_output.txt", "w") as f:
  for dic in scores_per_line_openai:
    for key in dic:
      f.write('%s\n' % (dic[key]['r']))
      break


# Extract only rouge-2 recall scores

counter = 0

with open("openai_rouge_2_recall_output.txt", "w") as f:
  for dic in scores_per_line_openai:
    for key in dic:
      counter += 1
      if (counter % 2 == 0):
        f.write('%s\n' % (dic[key]['r']))
        break

# Extract only rouge-l recall scores

counter_l = 0

with open("openai_rouge_l_recall_output.txt", "w") as f:
  for dic in scores_per_line_openai:
    for key in dic:
      counter += 1
      if (counter % 3 == 0):
        f.write('%s\n' % (dic[key]['r']))
        break

# Extract only rouge-1 precision scores

with open("openai_rouge_1_precision_output.txt", "w") as f:
  for dic in scores_per_line_openai:
    for key in dic:
      f.write('%s\n' % (dic[key]['p']))
      break


# Extract only rouge-2 precision scores

counter = 0

with open("openai_rouge_2_precision_output.txt", "w") as f:
  for dic in scores_per_line_openai:
    for key in dic:
      counter += 1
      if (counter % 2 == 0):
        f.write('%s\n' % (dic[key]['p']))
        break

# Extract only rouge-l precision scores

counter_l = 0

with open("openai_rouge_l_precision_output.txt", "w") as f:
  for dic in scores_per_line_openai:
    for key in dic:
      counter += 1
      if (counter % 3 == 0):
        f.write('%s\n' % (dic[key]['p']))
        break


# Extract only rouge-1 f1 scores

with open("openai_rouge_1_f_output.txt", "w") as f:
  for dic in scores_per_line_openai:
    for key in dic:
      f.write('%s\n' % (dic[key]['f']))
      break


# Extract only rouge-2 f1 scores

counter = 0

with open("openai_rouge_2_f_output.txt", "w") as f:
  for dic in scores_per_line_openai:
    for key in dic:
      counter += 1
      if (counter % 2 == 0):
        f.write('%s\n' % (dic[key]['f']))
        break

# Extract only rouge-l f1 scores

counter_l = 0

with open("openai_rouge_l_f_output.txt", "w") as f:
  for dic in scores_per_line_openai:
    for key in dic:
      counter += 1
      if (counter % 3 == 0):
        f.write('%s\n' % (dic[key]['f']))
        break


```
