```python
import pandas as pd
import csv
import json
```

```python
# nRowsRead = 1000
# df1 = pd.read_csv('utterances.csv', delimiter=',', nrows = nRowsRead)
# df1.dataframeName = 'utterances.csv'
filename = 'utterances-2sp.csv'
df = pd.read_csv(filename)
df.head

df2 = pd.read_csv("Simple Fine-tuning Results - Sheet1.csv")
df2.head
```

```python
list100 = []
for index in df2.index:
  list100.append(" " + df2.loc[index].Prompt + " " + df2.loc[index].Completion)

```

```python
# last_host = False
# found_prompt = False
new_data = []
prompt = ""
completion = ""
open_ai_format = False
lastID = 0


for index in df.index:
  if df.loc[index].is_host == False:
    prompt = " ".join((prompt, df.loc[index].utterance))
    lastID = df.loc[index].episode
  else:
    if lastID == df.loc[index].episode:
      completion = " ".join((completion, df.loc[index].utterance))
      # print(prompt+completion)
      try:
        if df.loc[index + 1].is_host == False:
          if completion.endswith("?"):
            if prompt + completion in list100:
              prompt = ""
              completion = ""
              # print("removed")
              continue
            else:
              new_data.append({"prompt": prompt, "completion": completion})
          prompt = ""
          completion = ""


      except ValueError:
         if completion.endswith("?"):

           if prompt + completion in list100:
             prompt = ""
             completion = ""
             continue
           else:
              new_data.append({"prompt": prompt, "completion": completion})
         prompt = ""
         completion = ""
      except KeyError:
         if completion.endswith("?"):
           if prompt + completion in list100:
             prompt = ""
             completion = ""
             continue
           else:
              new_data.append({"prompt": prompt, "completion": completion})
         prompt = ""
         completion = ""
        



```

```python
import random


random.seed(491)
```

```python
validation_set = []
for x in range(15000):
  index = random.randrange(len(new_data))
  validation_set.append(new_data[index])
```

```python
[new_data.remove(item) for item in new_data if item in validation_set]
```

```python

with open('train_question.json', 'w') as f:
  for line in new_data:
    json.dump(line, f)
    f.write('\n')

with open('valid_question.json', 'w') as f:
  for line in validation_set:
    json.dump(line, f)
    f.write('\n')
```
